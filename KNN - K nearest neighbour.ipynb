KNN

```python
def distance(x1 , x2):
    return np.sum((x1 - x2)**2)**0.5
```

```python
distance(np.array([2,2]),np.array([0,0]))
```

```python
def KNN(x,y,query,k = 5):
    m = x.shape[0]
    all_distance = []
    for i in range(m):
        d = distance(query,x[i])
        all_distance.append(d)
    all_distance = sorted(all_distance)
    return np.array(all_distance)[:k]
```

```python
KNN(x,y,query_point)
```
